```python
# feature engineering
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df.columns
```

```python
features = ["study_hours_per_day", "attendance_percentage", "sleep_hours", "part_time_job", "mental_health_rating",]
```

```python
target = "exam_score"
```

```python
df_model = df[features + [target]].copy()
```

```python
df_model
```

```python
le = LabelEncoder()
df_model["part_time_job"] = le.fit_transform(df_model["part_time_job"])
```

```python
df_model
```

```python
X = df_model[features]
y = df_model[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
models = {
    "Linear Regression":{ 
        "model":LinearRegression(),
        "params":{}
    },
    "Decision Tree": {
        "model": DecisionTreeRegressor(),
        "params": {"max_depth": [3, 5, 7, 10, None],"min_samples_split": [2, 5, 10],"min_samples_leaf": [1, 2, 4]}
    },
    "RandomForestRegressor":{
        "model": RandomForestRegressor(),
        "params": {"n_estimators": [50, 100, 200],"max_depth": [3, 5, 7, 10, None],"min_samples_split": [2, 5, 10],"min_samples_leaf": [1, 2, 4]}
    }
}
```

```python
best_models = []
```

```python
for name,config in models.items():
    print(f"Training {name}")

    grid = GridSearchCV(config['model'], config['params'], cv=5 , scoring='neg_mean_squared_error')
    grid.fit(X_train, y_train)

    y_pred = grid.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    best_models.append({
        "model": name,
        "best_estimator": grid.best_params_,
        "rmse": rmse,
        "r2": r2
    })
```

```python
best_models
```

```python
results_df = pd.DataFrame(best_models)
```

```python
results_df.sort_values(by="rmse")
```

```python
import joblib
```

```python
best_row = results_df.sort_values(by="rmse").iloc[0]
best_row
```

```python
best_model_name = best_row["model"]
best_model_name
```

```python
best_model_config = models[best_model_name] 
best_model_config
```

```python
final_model = best_model_config["model"].set_params(**best_row["best_estimator"])
final_model.fit(X_train, y_train)
```

```python
final_model.predict(X_test)
```

```python
# export the model
joblib.dump(final_model, "student_performance_best_model.pkl")
```
